Bokeh makes charts that you can zoom, hover, click, and explore — not just look at.

```python
# Deprecated Method: You were using p.circle(), which has been deprecated since Bokeh 3.4.0.
# Modern Replacement: The function was replaced by p.scatter(), which is more robust and specifically designed for scatter plots.
# By replacing the line:
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.iris import flowers

# It's good practice to ensure your input data types are clean
# Convert any potential object/URL columns in your 'flowers' data to strings if possible
# (The built-in sample data is usually fine, though)

output_file('test.html')

p = figure(title="test flower")
p.xaxis.axis_label = "petal length"
p.yaxis.axis_label = "petal width"

# Use the recommended 'scatter' method instead of 'circle'
p.scatter(x=flowers['petal_length'], y=flowers['petal_width'], size=8)

show(p)
```

```python

from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title = "line graph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.line(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python


x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("line2.html")
p = figure(title = "line plot")
p.line(x,y)
show(p)
```

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("line3.html")
p = figure(title = "line plot")
p.scatter(x,y,fill_color = 'red' , legend_label = 'red points',size = 20)
show(p)
```
